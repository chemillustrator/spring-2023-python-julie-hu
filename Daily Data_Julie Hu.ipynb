```python
import matplotlib.pyplot as plt
```

```python
food_inventory = {
'Poke bowl': 2,
'Soup': 4,
'Spicy crawfish fried rice': 2,
'Burger': 2,
'Ramen': 3,
'Sushi': 2,
'Ice cream': 1,
}
```

```python
plt.bar(range(len(food_inventory)), list(food_inventory.values()), align='center', color=['#f68084', '#fee140', '#fe5196', '#ff5858', '#fbed96', '#f3e7e9', '#c7eafd'])
plt.xticks(range(len(food_inventory)), list(food_inventory.keys()), rotation = 45)
plt.show()
```

```python
plt.scatter(range(len(food_inventory)), list(food_inventory.values()), align='center', color=['#f68084', '#fee140', '#fe5196', '#ff5858', '#fbed96', '#f3e7e9', '#c7eafd'])
plt.xticks(range(len(food_inventory)), list(food_inventory.keys()), rotation = 45)
plt.show()
```
